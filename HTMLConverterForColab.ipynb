```
from google.colab import files

file = files.upload()
file_name = list(file.keys())[0]
contents = file[file_name].decode()
```

```
import nbformat

notebook = nbformat.reads(contents, as_version=4)
```

```
# 1. Import the exporter
from nbconvert import HTMLExporter

# 2. Instantiate the exporter.
html_exporter = HTMLExporter()

# 3. Process the notebook we loaded earlier
(body, resources) = html_exporter.from_notebook_node(notebook)
```

```
import os

modified_file_name = os.path.splitext(file_name)[0] + '.html'
```

```
with open(modified_file_name, 'w') as file:
  file.write(body)

files.download(modified_file_name)
```
